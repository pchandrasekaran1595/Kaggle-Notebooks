### **Library Imports**

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from time import time
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")
```

### **Utilities**

```python
si_mean = SimpleImputer(missing_values=np.nan, strategy="mean")
sc_X = StandardScaler()


def breaker(num: int=50, char: str="*") -> None:
    print("\n" + num*char + "\n")

SEED = 42
```

### **Read Data**

```python
df = pd.read_csv("../input/tabular-playground-series-jul-2022/data.csv")
df.head(5)

X = df.iloc[:, 1:].copy().values
X = si_mean.fit_transform(X)
X = sc_X.fit_transform(X)
```

### **PCA Analysis**

```python
pca = PCA(n_components=2, random_state=SEED)

start_time = time()
new_X = pca.fit_transform(X)
sns.scatterplot(x=new_X[:, 0], y=new_X[:, 1], palette="Set1")
plt.grid()
plt.show()

breaker()
print(f"Time Taken : {(time()-start_time)/60:.2f} minutes")
breaker()
```

### **t-SNE Analysis**

```python
tsne = TSNE(n_components=2, random_state=SEED)

start_time = time()
new_X = tsne.fit_transform(X)
sns.scatterplot(x=new_X[:, 0], y=new_X[:, 1], palette="Set1")
plt.grid()
plt.show()

breaker()
print(f"Time Taken : {(time()-start_time)/60:.2f} minutes")
breaker()
```
