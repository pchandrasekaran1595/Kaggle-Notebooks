```python
import os
import random as r
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
SEED = 42
BASE_PATH = "../input/google-stock-price"


def breaker(num: int=50, char: str = "*") -> None:
    print("\n" + num*char + "\n")

    
def make_plot(df, columns: list) -> None:
    x_Axis = [i+1 for i in range(df.shape[0])]

    plt.figure(figsize=(18, 6))
    for i in range(len(columns)):
        plt.subplot(1, len(columns), i+1)
        plt.plot(x_Axis, df[columns[i]], "r", label=columns[i])
        plt.xlabel("Time")
        plt.ylabel("Price")
        plt.legend()
        plt.grid()
    plt.show()
```

```python
train_df = pd.read_csv(os.path.join(BASE_PATH, "Google_Stock_Price_Train.csv"))
test_df  = pd.read_csv(os.path.join(BASE_PATH, "Google_Stock_Price_Test.csv"))
```
