```python
import os
import cv2
import random as r
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from typing import Union
```

```python
def show_image(image: np.ndarray, title: Union[str, int, None]=None) -> None:
    plt.figure()
    plt.imshow(image, cmap="gray")
    plt.axis("off")
    if title: plt.title(title)
    plt.show()
```

```python
df = pd.read_csv("/kaggle/input/digit-recognizer-ii-2023/train.csv")
df.head()
```

```python
images = df.iloc[:, 1:].copy().values.reshape(-1, 28, 28).astype("uint8")
new_images_48 = np.array([cv2.resize(image, (48, 48), interpolation=cv2.INTER_AREA) for image in images])
# new_images = np.expand_dims(np.array([cv2.resize(image, (48, 48), interpolation=cv2.INTER_AREA) for image in x]), axis=3)

print("\n" + 50*"*" + "\n")
for i in range(10):
    show_image(new_images_48[i])

print("\n" + 50*"*" + "\n")
print(f"Mean : {new_images_48.mean() / 255:.5f}")
print(f"Std  : {new_images_48.std() / 255:.5f}")
print("\n" + 50*"*" + "\n")
```
