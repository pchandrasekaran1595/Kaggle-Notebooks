```python
import torch
from transformers import AutoFeatureExtractor, SwinForImageClassification, SwinModel, SwinConfig


names = [
    "swin-tiny-patch4-window7-224",
    "swin-small-patch4-window7-224",
    "swin-base-patch4-window7-224",
    "swin-base-patch4-window12-384",
]


for name in names:
    pretrained = AutoFeatureExtractor.from_pretrained("microsoft/" + name)
    model = SwinModel.from_pretrained("microsoft/" + name)
    
    torch.save(model.state_dict(), f"FE-{name}.pt")
```

```python
sizes = [ 224, 224, 224, 384 ]

num_heads = [ [3, 6, 12, 24], [3, 6, 12, 24], [4, 8, 16, 32], [4, 8, 16, 32], ]

depths = [ [2, 2, 6, 2], [2, 2, 18, 2], [2, 2, 18, 2], [2, 2, 18, 2], ]

embed_dims = [ 96, 96, 128, 128, ]

window_sizes = [ 7, 7, 7, 12, ]

for i in range(len(sizes)):
    config = SwinConfig(image_size=sizes[i], depths=depths[i], embed_dim=embed_dims[i], num_heads=num_heads[i], window_size=window_sizes[i])
    model = SwinModel(config)

    model.load_state_dict(torch.load(f"FE-{names[i]}.pt"))
```
